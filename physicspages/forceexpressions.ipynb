# Physical Quantities Expressed Using Force

Force is a fundamental physical quantity. Many other physical quantities are **defined directly or indirectly using force**. Below is a structured list with definitions, formulas, and SI units.

---

## 1. Work (W)

Work is done when a force causes displacement.

$$W = \vec{F} \cdot \vec{s} = Fs\cos\theta$$

**SI Unit:** Joule (J)

---

## 2. Power (P)

Power is the rate of doing work.

$$P = \frac{W}{t} = \vec{F} \cdot \vec{v}$$

**SI Unit:** Watt (W)

---

## 3. Impulse (J)

Impulse is the product of force and the time for which it acts.

$$\vec{J} = \vec{F}\Delta t = \Delta \vec{p}$$

**SI Unit:** Newton-second (N·s)

---

## 4. Pressure (P)

Pressure is force acting per unit area.

$$P = \frac{F}{A}$$

**SI Unit:** Pascal (Pa)

---

## 5. Acceleration (a)

From Newton’s Second Law of Motion:

$$\vec{F} = m\vec{a} \quad \Rightarrow \quad \vec{a} = \frac{\vec{F}}{m}$$

**SI Unit:** m/s²

---

## 6. Torque / Moment of Force (τ)

Torque is the turning effect of a force.

$$\vec{\tau} = \vec{r} \times \vec{F}$$

**SI Unit:** Newton-meter (N·m)

---

## 7. Stress (σ)

Stress is force applied per unit area of a material.

$$\sigma = \frac{F}{A}$$

**SI Unit:** Pascal (Pa)

---

## 8. Elastic Potential Energy / Strain Energy (U)

Energy stored in an elastic body due to deformation.

$$U = \frac{1}{2}Fx$$

**SI Unit:** Joule (J)

---

## 9. Centripetal Force Relation

Force required to keep a body in circular motion.

$$F = \frac{mv^2}{r}$$

This force determines:

- Radial acceleration
- Angular speed

---

## 10. Weight (W)

Weight is the gravitational force acting on a body.

$$W = mg$$

**SI Unit:** Newton (N)

---
