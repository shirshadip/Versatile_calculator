# Physical Quantities Expressed Using Force

Force is a fundamental physical quantity. Many other physical quantities are **defined directly or indirectly using force**. Below is a structured list with definitions, formulas, and SI units.

---

## 1. Work (W)

Work is done when a force causes displacement.

$$W = \vec{F} \cdot \vec{s} = Fs\cos\theta$$

**SI Unit:** Joule (J)

---

## 2. Power (P)

Power is the rate of doing work.

$$P = \frac{W}{t} = \vec{F} \cdot \vec{v}$$

**SI Unit:** Watt (W)

---

## 3. Impulse (J)

Impulse is the product of force and the time for which it acts.

$$\vec{J} = \vec{F}\Delta t = \Delta \vec{p}$$

**SI Unit:** Newton-second (N¬∑s)

---

## 4. Pressure (P)

Pressure is force acting per unit area.

$$P = \frac{F}{A}$$

**SI Unit:** Pascal (Pa)

---

## 5. Acceleration (a)

From Newton‚Äôs Second Law of Motion:

$$\vec{F} = m\vec{a} \quad \Rightarrow \quad \vec{a} = \frac{\vec{F}}{m}$$

**SI Unit:** m/s¬≤

---

## 6. Torque / Moment of Force (œÑ)

Torque is the turning effect of a force.

$$\vec{\tau} = \vec{r} \times \vec{F}$$

**SI Unit:** Newton-meter (N¬∑m)

---

## 7. Stress (œÉ)

Stress is force applied per unit area of a material.

$$\sigma = \frac{F}{A}$$

**SI Unit:** Pascal (Pa)

---

## 8. Elastic Potential Energy / Strain Energy (U)

Energy stored in an elastic body due to deformation.

$$U = \frac{1}{2}Fx$$

**SI Unit:** Joule (J)

---

## 9. Centripetal Force Relation

Force required to keep a body in circular motion.

$$F = \frac{mv^2}{r}$$

This force determines:

- Radial acceleration
- Angular speed

---

## 10. Weight (W)

Weight is the gravitational force acting on a body.

$$W = mg$$

**SI Unit:** Newton (N)

---

# üìå Force Formulas in Physics

This notebook lists the **most important force equations** used in classical physics, gravitation, electromagnetism, and modern physics.

---

## 1Ô∏è‚É£ Newtonian (Classical) Mechanics

### Newton‚Äôs Second Law

$$
\boxed{F = m a}
$$

Where:

- \(F\) = force (N)
- \(m\) = mass (kg)
- \(a\) = acceleration (m/s¬≤)

---

### Weight (Gravitational Force near Earth)

$$
\boxed{F = mg}
$$

- \(g \approx 9.8 \, \text{m/s}^2\)

---

### Frictional Force

$$
\boxed{F_f = \mu N}
$$

- \(\mu\) = coefficient of friction
- \(N\) = normal force

---

### Spring Force (Hooke‚Äôs Law)

$$
\boxed{F = -kx}
$$

- \(k\) = spring constant
- \(x\) = displacement

---

## 2Ô∏è‚É£ Universal Gravitation

### Newton‚Äôs Law of Gravitation

$$
\boxed{F = G \frac{m_1 m_2}{r^2}}
$$

- \(G = 6.674 \times 10^{-11} \, \text{N¬∑m}^2/\text{kg}^2\)
- \(r\) = distance between masses

---

## 3Ô∏è‚É£ Electromagnetic Forces

### Coulomb‚Äôs Law (Electric Force)

$$
\boxed{F = k \frac{q_1 q_2}{r^2}}
$$

- \(k = \frac{1}{4\pi\varepsilon_0}\)
- \(q_1, q_2\) = charges

---

### Electric Force in an Electric Field

$$
\boxed{F = qE}
$$

- \(E\) = electric field

---

### Magnetic Force on Moving Charge

$$
\boxed{F = qvB\sin\theta}
$$

---

### Lorentz Force (Electric + Magnetic)

$$
\boxed{\vec{F} = q(\vec{E} + \vec{v} \times \vec{B})}
$$

---

### Force on Current-Carrying Conductor

$$
\boxed{F = BIL\sin\theta}
$$

---

## 4Ô∏è‚É£ Circular Motion

### Centripetal Force

$$
\boxed{F = \frac{mv^2}{r}}
$$

or

$$
\boxed{F = m\omega^2 r}
$$

---

## 5Ô∏è‚É£ Momentum and Impulse

### Force as Rate of Change of Momentum

$$
\boxed{F = \frac{dp}{dt}}
$$

---

### Impulse

$$
\boxed{J = F\Delta t}
$$

---

## 6Ô∏è‚É£ Fluid Mechanics

### Buoyant Force (Archimedes‚Äô Principle)

$$
\boxed{F_b = \rho V g}
$$

- \(\rho\) = density of fluid
- \(V\) = volume displaced

---

### Drag Force (Approximate)

$$
\boxed{F_d = \frac{1}{2} C_d \rho A v^2}
$$

---

## 7Ô∏è‚É£ Relativity & Modern Physics

### Relativistic Force (Conceptual)

$$
\boxed{F = \frac{dp}{dt}, \quad p = \gamma mv}
$$

---

## 8Ô∏è‚É£ Nuclear & Fundamental Forces (Conceptual)

### Strong Nuclear Force

$$
\boxed{F \propto \frac{1}{r^2} \quad (\text{very short range})}
$$

---

### Weak Nuclear Force

$$
\boxed{F \ll \text{Strong Force}, \text{short range}}
$$
